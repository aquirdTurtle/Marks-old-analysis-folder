```python
analog_on = srs(analogSRS_DataHome + "Analog On.TXT")
analog_off = srs(analogSRS_DataHome + "Analog Off.TXT")

# -4dB gain 14kHz corner
init_digital_on = srs(digitalSRS_DataHome + "Digital On.TXT")

# -9dB gain 14kHz corner
digital_on_low_gain = srs(digitalSRS_DataHome + "Digital On Low Gain.TXT")

# -9dB gain 30kHz corner
digital_on_low_gain_high_corner = srs(digitalSRS_DataHome + "Digital On Low Gain High Corner.TXT")

# -4dB gain 30kHz corner
digital_4 = srs(digitalSRS_DataHome + "Digital On High Gain High Corner.TXT")

# -4dB gain 14kHz corner 5kHz low pass corner
digital_5 = srs(digitalSRS_DataHome + "Digital On Low Low Pass Corner.TXT")

# -4dB gain 14kHz corner 5kHz low pass corner
digital_6 = srs(digitalSRS_DataHome + "Digital On High Low-Pass Corner.TXT")
```
